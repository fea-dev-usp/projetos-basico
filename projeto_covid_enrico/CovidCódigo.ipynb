```python
covidpibcapita=covid_dados[['location','gdp_per_capita']]
covidpibcapita.set_index('location', inplace=True)
covidpibcapita = covidpibcapita.drop('World',axis=0)
covidpibcapita = covidpibcapita.drop_duplicates()
covidpibcapita
```

```python
covidmortecapita=covid_dados[['location','total_deaths_per_million']]
covidmortecapita = covidmortecapita.drop_duplicates(subset=['location'],keep='last')
covidmortecapita.set_index('location', inplace=True)
covidmortecapita = covidmortecapita.drop(['World','International'],axis=0)

covidmortecapita
```

```python
#df1.join(df2, how='outer')
covidnovo=covidpibcapita.join(covidmortecapita,how='outer')
#plt.scatter(covidpibcapita.values,covidpaismorte.values)
#plt.show
covidnovo
```
